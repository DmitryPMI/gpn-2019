# №1

<p> Начну с того, что я зачем-то спрогнозировал спрос по городам, не заметив, что он уже был дан в задаче на 3 месяца вперед. С одной стороны это может говорить о моей невнимательности, но с другой стороны,  реальной задаче его пришлось бы прогнозировать. </p>
<br>
<p> Итак, первая задача - прогнозирование временных рядов. Большинство из рядов стационарные, а значит для них можно подобрать модель в классе ARIMA. </p>
<p> Однако данная задача - задача множественного прогнозирования. Подбирать параметры ARIMA для 160+ рядов достаточно долго и энергозатратно. Я пробовал для некоторых, но для всех это долго.</p>
<p> Здесь хорошим решением может стать autoarima, которая реализована в R и имеет аналог в python. </p>
<br>
<p> Тем не менее, для задач множественного прогнозирования часто используют другие методы, например градиентный бустинг или линейную регрессию с регуляризаторами. Я использовал xgboost. </p>

<p> Для локации VL нет данных по спросу, так же ряды для некоторых sku даны не до июня 2019 года, это усложнаяет задачу прогнозирвания. В первом случае можно убрать спрос из признаков, во втором точно вообще ничего сказать нельзя. Можно сделать наивный прогоз или взять, например, медиану. <p>

<p> До дедлайна меньше 5 часов и я ругаю себя, что не потратил больше времени заранее, но это достаточно распространенная проблема. Если бы у меня было больше времени, я бы, наверное, попробовал применить в этой задаче глубокое обучение, или нашел более емкий способ использовать авторегрессию. </p>

# №2

Если использовать бустинг или линейную регрессию с регуляризаторами, то близкую к оптимальной цену можно найти достаточно просто. Я использовал цену как один из факторов, которые влияют на объем продаж. Я предсказывал объемы продаж для разных цен с шагом 0.1. Такой способ дает оптимальную цену с погрешностью 10%. 

В случае авторегрессии нужно нужно учитывать параметр $ \lambda$ , который влияет на прогнозный интервал. Это даст большую погрешность. 

Задача поиска оптимальной цены - реальная и интересная задача. Тут нужно учитывать, что мы не можем сильно менять цену каждый месяц. Покупатели просто этого не поймут, а с крупными покупателями есть контракты. Так же нужно учитывать корреляции между товарами в каждой локации. 
В такой формулировке задача уже становится сложной, но ее можно сформулировать в точной постановке и решить каким-нибудь эвристичесим алгоритмом.
